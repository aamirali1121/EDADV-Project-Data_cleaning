```python

import pandas as pd

df = pd.read_csv('amazon.csv')
df_raw = df.copy()
df.head()
```

```python
df = df.rename(columns={
    "a-size-base-plus": "brand",
    "a-size-base-plus 2": "title",
    "a-price-whole": "price",
    "a-offscreen": "mrp",
    "a-offscreen 2": "discount_percent",
    "a-icon-alt": "rating_text",
    "a-size-mini": "rating_count",
    "s-image src": "image_url",
    "a-link-normal href": "product_url"
})

df.head()
```

```python
import numpy as np

# PRICE
df["price"] = pd.to_numeric(df["price"].astype(str).str.replace(",", ""), errors="coerce")

# MRP
df["mrp"] = df["mrp"].astype(str).str.replace("₹", "").str.replace(",", "")
df["mrp"] = pd.to_numeric(df["mrp"], errors="coerce")

# DISCOUNT PERCENT
df["discount_percent"] = df["discount_percent"].astype(str).str.extract(r'(\d+)').astype(float)

# RATING (Extract number)
df["rating"] = df["rating_text"].astype(str).str.extract(r'([\d\.]+)').astype(float)

# RATING COUNT
df["rating_count"] = df["rating_count"].astype(str).str.replace(",", "", regex=False)
df["rating_count"] = pd.to_numeric(df["rating_count"], errors="coerce")
```

```python
df_clean = df[[
    "brand", "title", "price", "mrp", "discount_percent",
    "rating", "rating_count", "product_url", "image_url"
]]

df_clean.head()
```

```python
df_clean = df_clean.dropna(subset=["price"])   # price is required
df_clean["rating"] = df_clean["rating"].fillna(df_clean["rating"].median())
df_clean["rating_count"] = df_clean["rating_count"].fillna(0)
df_clean["discount_percent"] = df_clean["discount_percent"].fillna(0)
df_clean["mrp"] = df_clean["mrp"].fillna(df_clean["price"])
```

```python
import matplotlib.pyplot as plt
import numpy as np

prices = df_clean["price"].dropna()

plt.figure(figsize=(10,6))
plt.hist(prices, bins=40, edgecolor='black', alpha=0.7)

plt.title("Price Distribution with Summary Statistics", fontsize=14)
plt.xlabel("Price (₹)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(alpha=0.3)

# Summary statistics text box
mean_val = prices.mean()
median_val = prices.median()
std_val = prices.std()

stats_text = f"Mean: ₹{mean_val:.2f}\nMedian: ₹{median_val:.2f}\nStd Dev: ₹{std_val:.2f}"

plt.text(0.75, 0.75, stats_text, transform=plt.gca().transAxes,
         fontsize=12, bbox=dict(facecolor='white', alpha=0.6))

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df_clean["rating"], bins=20, edgecolor="black")
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.grid(alpha=0.2)
plt.show()
```

```python
discounts = df_clean["discount_percent"].dropna()

plt.figure(figsize=(10,6))
plt.hist(discounts, bins=30, edgecolor="black", alpha=0.7)

plt.title("Distribution of Discount Percent (with Summary Statistics)", fontsize=14)
plt.xlabel("Discount (%)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(alpha=0.3)

# Summary Statistics
mean_val = discounts.mean()
median_val = discounts.median()
std_val = discounts.std()

stats_text = f"Mean: {mean_val:.2f}%\nMedian: {median_val:.2f}%\nStd Dev: {std_val:.2f}%"

plt.text(0.73, 0.75, stats_text, transform=plt.gca().transAxes,
         fontsize=12, bbox=dict(facecolor='white', alpha=0.7))

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df_clean["price"], df_clean["rating"], alpha=0.5)

plt.title("Price vs Rating (with Insight Box)", fontsize=14)
plt.xlabel("Price (₹)", fontsize=12)
plt.ylabel("Rating", fontsize=12)
plt.grid(alpha=0.3)

info_text = (
    "Observations:\n"
    "- No strong relationship\n"
    "- Cheap and expensive watches get similar ratings\n"
    "- Low-priced items show slightly more low ratings"
)

plt.text(0.65, 0.15, info_text, transform=plt.gca().transAxes,
         fontsize=11, bbox=dict(facecolor='white', alpha=0.7))

plt.tight_layout()
plt.show()
```

```python
brand_counts = df_clean["brand"].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.barh(brand_counts.index, brand_counts.values, color="steelblue")

plt.title("Top Brands by Listing Count", fontsize=14)
plt.xlabel("Count", fontsize=12)
plt.ylabel("Brand", fontsize=12)
plt.gca().invert_yaxis()
plt.grid(alpha=0.3)

info = (
    f"Top Brand: {brand_counts.index[0]}\n"
    f"Listings: {brand_counts.iloc[0]}"
)

plt.text(0.65, 0.8, info, transform=plt.gca().transAxes,
         fontsize=11, bbox=dict(facecolor='white', alpha=0.7))

plt.tight_layout()
plt.show()
```

```python
avg_rating = df_clean.groupby("brand")["rating"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(avg_rating.index, avg_rating.values, color="teal")

plt.title("Average Rating by Brand (Top 10)", fontsize=14)
plt.xlabel("Average Rating", fontsize=12)
plt.ylabel("Brand", fontsize=12)
plt.gca().invert_yaxis()
plt.grid(alpha=0.3)

top_brand = avg_rating.index[0]
top_rating = avg_rating.iloc[0]

info = f"Highest Rated Brand:\n{top_brand} ({top_rating:.2f}★)"

plt.text(0.65, 0.75, info, transform=plt.gca().transAxes,
         fontsize=11, bbox=dict(facecolor='white', alpha=0.7))

plt.tight_layout()
plt.show()
```
